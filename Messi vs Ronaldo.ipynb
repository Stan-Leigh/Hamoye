### Cleaning Messi's Career stats Data

```python
df.info()
```

```python
df['Played_in'] = df['Club']['Club']
df['Seasons'] = df['Season']['Season']
df['Division/League'] = df['League']['Division']
df['National Cup_Apps'] = df['National Cup']['Apps']
df['National Cup_Goals'] = df['National Cup']['Goals']
df['Europe[a]_Apps'] = df['Europe[a]']['Apps']
df['Europe[a]_Goals'] = df['Europe[a]']['Goals']
df['Other_Apps'] = df['Other']['Apps']
df['Other_Goals'] = df['Other']['Goals']
df['Total_Apps'] = df['Total']['Apps']
df['Total_Goals'] = df['Total']['Goals']
```

```python
df.drop(columns=['Club', 'Season', 'League', 'National Cup', 'Europe[a]', 'Other', 'Total'], inplace=True)
```

```python
df
```

```python
df.drop(index=[3, 21], inplace=True)
```

```python
df
```

```python
df.info()
```

```python
df.iloc[1, 4]
```

```python
df.columns
```

```python
# Removing the second value empty string in each column tuple.
new_col = []
for column in df.columns:
    new = column[0]
    new_col.append(new)
    
```

```python
df.columns = new_col
```

```python
df.columns
```

```python
df
```

```python
df.reset_index(inplace=True) # Resetting the indexes for easier looping
```

```python
df
```

```python
df.drop(columns=['index'], inplace=True)
```

```python
df
```

```python
# Replace the dashes with zeroes
for column in range(3, 11):
    for row in range(0, 22):
        if df.iloc[row, column] == '—':
            df.iloc[row, column] = 0
```

```python
df
```

```python
# Convert to integer type
df = df.astype({'National Cup_Apps': 'int64', 'National Cup_Goals': 'int64', 'Europe[a]_Apps': 'int64',
           'Europe[a]_Goals': 'int64', 'Other_Goals': 'int64', 'Total_Apps': 'int64', 
           'Total_Goals': 'int64'})
```

```python
df.info()
```

```python
# Use regex expressions to remove the non numeric characters beside the numeric characters.
df['Other_Apps'] = df['Other_Apps'].str.extract(r'(\d)')
```

```python
df
```

```python
# Fill Nan values wih zero
df['Other_Apps'].fillna(0, inplace=True)
```

```python
df = df.astype({'Other_Apps': 'int64'})
```

```python
df.info()
```

```python
df
```
